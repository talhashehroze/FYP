```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
twitter_username = 'eorta84'
tweet_limit = 0
```

```python
tweets_list = []
```

```python
for i, tweet in enumerate(
        sntwitter.TwitterUserScraper(twitter_username).get_items()):

    if tweet.user.profileImageUrl.startswith(
            "https://abs.twimg.com/sticky/default_profile_images/"):
        xdefault_profile_image = 'TRUE'
    else:
        xdefault_profile_image = 'FALSE'

    # # custom logic
    if xdefault_profile_image == 'FALSE' or tweet.user.profileBannerUrl or tweet.user.renderedDescription or tweet.user.verified or tweet.user.location or tweet.user.link:
        xdefaultProfile = 'FALSE'
    else:
        xdefaultProfile = 'TRUE'

    if i > tweet_limit:
        break
    tweets_list.append([
        tweet.user.created,
        xdefaultProfile,
        xdefault_profile_image,
        tweet.user.renderedDescription,
        tweet.user.favouritesCount,
        tweet.user.followersCount,
        tweet.user.friendsCount,
        # tweet.user.geo_enabled,
        tweet.user.id,
        tweet.lang,
        tweet.user.location,
        tweet.user.profileBannerUrl,
        tweet.user.profileImageUrl,
        tweet.user.displayname,
        tweet.user.statusesCount,
        tweet.user.verified,
        tweet.user.statusesCount /
        (pd.Timestamp.now().date() - tweet.user.created.date()).days,
        (pd.Timestamp.now().date() - tweet.user.created.date()).days
    ])
```

```python
tweets_df = pd.DataFrame(
    tweets_list,
    columns=[
        'created_at',
        'default_profile',
        'default_profile_image',
        'description',
        'favourites_count',
        'followers_count',
        'friends_count',
        # 'geo_enabled',
        'id',
        'lang',
        'location',
        'profile_background_image_url',
        'profile_image_url',
        'screen_name',
        'statuses_count',
        'verified',
        'average_tweets_per_day',
        'account_age_days'
    ])
```

```python
tweets_df.to_csv('custom_twitter_human_bots_dataset.csv', index=False)
```
